# Pre-Processing

## Training Data

### Import data

```python
import pandas as pd
import numpy as np
dataSet = pd.read_csv('E:\Docx\Projects\Predictive Maintenance\dataset.csv',encoding="ISO-8859-1")
```

```python
DataSet = DataSet.drop(['Unnamed: 0'],axis = 1)
```

### Generating RUL,Label1,Label2.

```python
group = dataSet.groupby(['id'])
```

```python
max_cycle = group.cycle.max()
#print(max_cy)
```

```python
cycle_frame = max_cycle.to_frame()
cycle_frame = cycle_frame.assign(id = pd.Series(range(0,101)))
#print(cy_frame)
```

```python
dataSet = dataSet.merge(cycle_frame,left_on='id',right_on='id',how='outer')
#print(dataset)
#list(dataSet)
```

```python
dataSet['cycle_y'] = dataSet['cycle_y'].sub(dataSet['cycle_x'],axis = 0)
#list(dataSet)
```

```python
dataSet.rename(index = str, columns={"cycle_y":"RUL"},inplace=True)
```

```python
dataSet.rename(index = str, columns={"cycle_x":"cycle"},inplace=True)
```

```python
# w0,w1 - No. of cycles within which system failure is expected. (Constants)
w1 = 30
w0 = 15
dataSet['label1'] = np.where(dataSet['RUL']<=w1, 1, 0)
dataSet['label2'] = np.where(dataSet['RUL']<=w0,2,np.where(dataSet['RUL']<=w1,1,0))
#print(dataSet)
```

### Normalizing the sensor data

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
dataSet[['s1','s2','s3','s4','s5','s6','s7','s8','s9','s10','s11','s12','s13','s14','s15','s16','s17','s18','s19','s20','s21']] = scaler.fit_transform(dataSet[['s1','s2','s3','s4','s5','s6','s7','s8','s9','s10','s11','s12','s13','s14','s15','s16','s17','s18','s19','s20','s21']])
```
